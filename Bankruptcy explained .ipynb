## 1-Introduction

Cette étude traitera la prédiction de faillite d'une entreprise. Le but de la prédiction de faillite est de développer un modèle prédictif qui combine différentes mesures économétriques et permet de prévoir la situation financière d'une entreprise.

C'est un vaste domaine de la finance et de l'économétrie qui combine les connaissances d'experts sur le phénomène et les données historiques des entreprises prospères et non prospères.

Généralement, les entreprises sont quantifiées par de nombreux indicateurs qui décrivent leur situation commerciale et qui sont ensuite utilisés pour induire un modèle mathématique à partir d'observations passées.

Différentes questions sont associées à la prévision de faillite. Deux problèmes principaux sont les suivants. Tout d'abord, les indicateurs économétriques décrivant l'état de l'entreprise sont proposés par des experts du domaine. Cependant, il n'est pas très clair comment les combiner en un puissant modèle. Deuxièmement, les observations historiques utilisées pour former un modèle sont habituellement influencées par le phénomène de déséquilibre des données, parce qu'il y a généralement beaucoup plus d'entreprises qui réussissent que celles qui sont en faillite. Par conséquent, le modèle formé tend à prédire que les entreprises réussies (classe majoritaire), même lorsque certaines d'entre elles sont des entreprises en difficulté. Ces deux questions influent surtout sur le processus de la la capacité de prédiction finale du modèle.

## 4-Nettoyage et prétraitement des données

Après avoir visualisé et étudié la distribution des caractéristiques, nous en sommes arrivés à la conclusion :

- Présence de NaNs dans l'ensemble de données qui nécessitent un certain remplacement.
    - Nous avons remplacé les nans par la médiane puisque nous avons beaucoup de valeurs aberrantes. 
- Présence de types de données binaires
    - Effectuer un casting sur les données pour changer le type
- Les échelles varient beaucoup d'une caractéristique à l'autre, d'où la normalisation.
    - Nous normalisons avec la formule : $ X^* = \frac{X - \bar{X}}{\max X - \min X }$
- Présence de valeurs aberrantes
    - Supprimer les valeurs aberrantes (seulement pour les entreprises qui n'ont pas fait de faillite) qui vérifient $|Y| > \alpha|\bar{Y}|$ étant donné que $\alpha \sim 70 \text{ (expérimentalement)}$.

Nous allons créer une nouvelle dataset nettoyée.



```python
def process(year, df):
    # Fill NAN values with median
    def fillna_median(c):
        return c.fillna(c.median())

    # Normalizing data 
    def normalize(series):
        if series.name != "class":
            return (series - series.mean())/ (series.max() - series.min())
        else:
            return series
    
    def f(p):
        def outliers(series):
            if series.name != "class":
                mean = np.abs(series.mean())
                series[np.abs(series) > p * mean] = np.nan
                return series
            else:
                return series
        return outliers


    log.info("year {}: Filling NaNs with median ...".format(year))
    df = df.apply(fillna_median, axis=0)
    log.info("year {}: NaNs filled !".format(year))

    log.info("year {}: Dropping outliers data ...".format(year))
    df = df.apply(f(80), axis=0)
    df = df.dropna()
    log.info("year {}: Outliers dropped !".format(year))
    
    log.info("year {}: Normalizing data ...".format(year))
    df = df.apply(normalize, axis=0)
    log.info("year {}: Data normalized !".format(year))


    
    return df

log.info("Processing and saving cleaned datasets...")
pp_dfs = {}
for key, value in dfs.items():
    pp_dfs[key] = process(key, value)
    # Save File
    pp_dfs[key].to_csv('data/cleaned-dataset-{}.csv'.format(key), index=False)

log.info("Data processed and saved!")
```

### Réduction de dimension

```python
# Calculate ACP, and returns the number of ACP components to have percent% of variance explained

def acp(df, percent, plot_cumm = True):
    S = np.diag(1/df.iloc[:,:-1].std())
    mu = np.matrix(df.iloc[:,:-1].mean())
    U = np.matrix([np.ones(df.iloc[:,:-1].shape[0])])
    X = pd.DataFrame((df.iloc[:,:-1] - U.T.dot(mu)).dot(S).values,columns = df.columns[:-1])

    Var_X = X.cov()
    val_propres,vec_propres = np.linalg.eig(Var_X)
    ordre = val_propres.argsort()[::-1]
    val_propres = val_propres[ordre]
    vec_propres = vec_propres[:,ordre]

    row = ["Comp." + str(i+1) for i in range(64)]
    col = ["Std deviation","Prop.of Variance(%)","Cumulative Prop.(%)"]
    tot = sum(val_propres)
    res_1 = pd.DataFrame([[i**0.5,np.round(i*100 /tot,2),np.round(j*100/tot,2)]
                          for i,j in zip(val_propres,np.cumsum(val_propres))],
                            columns= col,index = row)
    res_2 = pd.DataFrame(vec_propres,columns = row,
                                     index = df.columns[:-1])
    if plot_cumm:
        plt.clf()
        plt.subplot(211)
        plt.axis([0.5,21.5,0,102])
        plt.bar(range(1,65), res_1.iloc[:,1], alpha=0.5,
                     align='center',
                    label="Proportion of Variance(%)")
        plt.step(range(1,65),res_1.iloc[:,2], where='mid',
                     label="Cumulative Proportion(%)")
        plt.ylabel("Variance explained")
        plt.xlabel("Principal component")
        plt.figure(figsize=(18, 4))
        plt.legend(loc='best')
        plt.subplot(212)
        plt.axis([0.5,20.5,0,40])
        plt.plot(range(1,65),res_1.iloc[:,1], linestyle='--',
                      marker='o', color='black')
        plt.ylabel("Eigenvalue")
        plt.xlabel("Principal component")
        plt.title("Scree plot")
        plt.figure(figsize=(18, 4))
        plt.tight_layout()
    return int(res_1.index[res_1["Cumulative Prop.(%)"] > percent].tolist()[0].split('.')[-1])
```

```python
from sklearn.decomposition import PCA as princomp
def reduce_dim(df_, n):
    cpa = princomp(n_components=n)
    Z = cpa.fit_transform(df_)
    Z = np.concatenate([Z, df_['class'].as_matrix().reshape(df_.shape[0], 1)], axis=1)
    Zdf_ = pd.DataFrame(Z).rename({n: 'class'}, axis='columns')
    return Zdf_
```

## 6-Evaluation des performances

### Note
En raison de la nature déséquilibrée des données de formation, nous utiliserons le critère ROC-AUC pour évaluer la qualité des modèles. Pour chacun des modèles considérés, nous examinerons la qualité de la prédiction à l'aide d'une méthodologie de validation croisée à 10 plis.

Ici, on définit une classe qui permet de tester plusieurs algorithmes de classification, d'utiliser la validation croisée pour les évaluer, et enfin de prédire à l'aide du modèle pré-entrainé.

```python
class Classifier:
    def __init__(self, train_df, test_df=None, model=None, **kwargs):
        self.df = train_df
        self.tdf = test_df
        if type(model) == type('') or not model:
            if (model == "linear") or model == "l":
                self.model = linear_model.SGDClassifier(**kwargs)
            elif (model == "random_forst") or model == "rf":
                self.model = ensemble.RandomForestClassifier(**kwargs)
            elif model == "svm":
                self.model = svm.SVC(**kwargs)
            elif model == "nn":
                self.model = neural_network.MLPClassifier(**kwargs)
            elif model == "knn":
                self.model = neighbors.KNeighborsClassifier(**kwargs)
            elif model == "xgb":
                self.model = xgboost.XGBClassifier(**kwargs)
            elif model == 'gb':
                params = {'n_estimators': 500, 'max_depth': 4, 'min_samples_split': 2,
                          'learning_rate': 0.01}
                self.model = ensemble.GradientBoostingClassifier(**params)
            self.fit = True
        else:
            self.fit = False
            self.model = model

        
    def cross_validate(self, test_size=0.2, fit=True):
        X = self.df.drop(labels=(['class']), axis=1).as_matrix()
        y = self.df['class'].as_matrix()
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=0)
        if fit and self.fit:
            self.model.fit(X_train, y_train)
            pickle.dump(self.model, open("model.bin", 'wb'))
            self.fit = False
        y_pred = self.model.predict(X_test)
        return roc_auc_score(y_test, y_pred)

    
    def predict(self):
        X = self.df.drop(labels=(['class']), axis=1).as_matrix()
        y = self.df['class'].as_matrix()
        rX = self.tdf.drop(labels=['class'], axis=1).as_matrix()
        if self.fit:
            self.model.fit(X, y)
            pickle.dump(self.model, open("model.bin", 'wb'))
        y_pred = self.model.predict(rX)
        return y_pred
```

### Prédiction sur la même année

```python
models = ["linear", "random_forst", "svm", "knn", "xgb"]
modelss = [3, 5, 7, 1, 6]
cross_val = []
roc_auc = []
year = 4
for m in models:
    c = Classifier(train_df=dr_dfs[year], test_df=dr_dfs[year], model=m)
    cross_val.append(np.mean(cross_val_score(c.model, c.tdf.drop(labels=['class'], axis=1), c.tdf['class'].as_matrix(), cv=10)))
    roc_auc.append(roc_auc_score(c.predict(), c.tdf['class'].as_matrix()))
plt.figure(figsize=(18, 4))
plt.bar(models, height= cross_val, color="firebrick")
plt.title("Cross-validation of 10 folds for year " + str(year))
for a,b in zip(models, cross_val):
        plt.text(a, b, str(b))
plt.figure(figsize=(18, 4))
plt.bar(models, height= roc_auc, color="powderblue")
plt.title("ROC_AUC for year " + str(year))
for a,b in zip(models, roc_auc):
        plt.text(a, b, str(b))
```

nous constatons que les cinq algorithmes parviennent à prédire parfaitement pour les données de la même année avec une précision égal à 100%.

### Test hors échantillon

Ici, nous allons former le modèle sur une année et l'utiliser pour prédire les autres années.
Ex: on va entrainer le modèle sur les données de l'année 1 et on va prédire sur les autres années.

```python
def pred_from_year(ly, py):
    models = ["linear", "random_forst", "svm", "knn", "xgb"]
    modelss = [3, 5, 7, 1, 6]
    cross_val = []
    roc_auc = []
    for m in models:
        c = Classifier(train_df=dr_dfs[ly], test_df=dr_dfs[py], model=m)
        cross_val.append(np.mean(cross_val_score(c.model, c.tdf.drop(labels=['class'], axis=1), c.tdf['class'].as_matrix(), cv=10)))
        roc_auc.append(roc_auc_score(c.predict(), c.tdf['class'].as_matrix()))
    plt.figure(figsize=(18, 4))
    plt.bar(models, cross_val, color="firebrick")
    plt.title("Cross-validation of 10 folds for year " + str(py))
    for a,b in zip(models, cross_val):
        plt.text(a, 1, str(b))
    plt.figure(figsize=(18, 4))
    plt.bar(models, roc_auc, color="powderblue")
    plt.title("ROC_AUC for year " + str(py))
    for a,b in zip(models, roc_auc):
        plt.text(a, 1, str(b))
learning_year = 1
predicted_years = range(2,6)
for y in predicted_years:
    pred_from_year(learning_year, y)
```

On remarque que mis à part la quatrième année les précisions **AUC** sont excellentes, donc notre modèle se généralise bien sur des données en dehors de l'échantillon de l'apprentissage. Et la validation et presque toujours parfaite. En effet le critère AUC et le plus adapté pour notre cas puisque on doit metrre l'accent aussi sur les faux positifs et les faux negatifs que sur la somme des deux.
